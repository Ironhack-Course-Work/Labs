```python
# import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://web.archive.org/web/20080619011029/http://www.rollingstone.com/news/coverstory/500songs/page/1"
```

```python
response = requests.get(url)
response.status_code
```

```python
#create the soup 

soup = BeautifulSoup(response.content, parser="html5lib")

soup
```

```python
#iterate through all 5 webpages

iterations = range(1,6)

for i in iterations:
    start_at= str(i)
    url = "https://web.archive.org/web/20080619011029/http://www.rollingstone.com/news/coverstory/500songs/page/" + start_at
    print(url)
```

```python
#Assembling the script to send and store multiple requests

pages = []

for i in iterations:
    # assemble the url:
    start_at= str(i)
    url = "https://web.archive.org/web/20080619011029/http://www.rollingstone.com/news/coverstory/500songs/page/" + start_at

    # download html with a get request:
    response = requests.get(url)

    # monitor the process by printing the status code
    print("Status code: " + str(response.status_code))

    # store response into "pages" list
    pages.append(response)
```

```python
#Checking the content of the 5 pages with index

BeautifulSoup(pages[0].content, "html.parser")
```

```python
#Because the first entry of every page is not in a <p> tag, we need to manually find it

soup.find("div", class_="text").find("a").text.replace("\n","")
```

```python
#finding all the song titles

soup.find_all("em")
```

```python
#cleaning all the songs and putting them in a list

#songs_clean = songs.replace("<em>","")

songs = []
for i in soup.find_all("em"):
    
    songs.append(i.get_text().replace("\n", " ").lstrip(" "))
```

```python
#Putting all the elements together to create bot lists (Thanks Flo!)


pages_parsed = []
songs500 = []
artists500 = []

for i in range(len(pages)):
    pages_parsed.append(BeautifulSoup(pages[i].content, "html.parser"))
    # select only the info about the songs
    songs_html = pages_parsed[i].find_all("em")
    for j in songs_html:
        songs500.append(j.get_text().replace("\n", " ").lstrip(" "))
    artists_raw = pages_parsed[i].select("div.coverstory-specialcover > div.text > p")[:-2]
    artists500.append([soup.find("div", class_="text").find("a").text.replace("\n","")] + [p.text.rsplit(",",1)[1].replace("\n", " ").lstrip(" ") for p in artists_raw])
    
```
